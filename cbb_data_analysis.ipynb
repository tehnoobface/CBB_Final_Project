```python
# Read in cbb.csv:
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

cbb = pd.read_csv('cbb.csv')
```

```python
# make a new column called "SEASON_FINAL" that takes the column "POSTSEASON" and assigns a value: Champion = 1, 2nd = 2, etc.:
cbb['SEASON_FINAL'] = cbb['POSTSEASON'].replace(['Champions', '2ND', 'F4', 'E8', 'S16', 'R32', 'R64', 'R68', 'NaN'], [1, 2, 4, 8, 16, 32, 64, 68, 0])


# Sort by "SEASON_FINAL" and then by "YEAR":
cbb = cbb.sort_values(['YEAR', 'SEASON_FINAL'], ascending=[True, True])


# Count number of NaN and NA values in SEASON_FINAL column:
cbb['SEASON_FINAL'].isna().sum()
# Makee NaN values in SEASON_FINAL column equal to 0:
cbb['SEASON_FINAL'] = cbb['SEASON_FINAL'].fillna(0)

cbb['SEASON_FINAL'].isna().sum()

# show value count for "SEASON_FINAL" column:
cbb['SEASON_FINAL'].value_counts()

cbb.head(10)
```

```python
# Count number of teams per year:
cbb['YEAR'].value_counts().sort_index()
```
